```python
df = pd.read_csv('../../data/stars_classification_with_images.csv')
```

```python
def get_photos(coordinates : list):
    coor_list = []
    for alpha, delta in coordinates:
        coor_list.append(alpha+delta+'\r\n')

    data = {
    'radecs': coor_list,
    }
    response = requests.post('https://dr12.sdss.org/bulkFields/raDecs', headers=headers, data=data)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all image tags in the table
    image_tags = soup.select('table#fieldTable img')

    # Extract the src attribute of each image tag
    image_urls = [img['src'] for img in image_tags]
    # Equivalent
    # image_urls = []
    # for img in image_tags:
    #     image_urls.append(img['src'])

    # Print the extracted image URLs
    return 'https://dr12.sdss.org/'+ str(image_urls[-1])
```

```python
get_photos([df['alpha'][0], df['delta'][0]])
```
